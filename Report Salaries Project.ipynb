### PROJECT DATA JOB SALARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
salaries = pd.read_csv(r'E:\Project\Project 2 - Job DA\salaries.csv')
```

```python
salaries
```

```python
# work_year
# experience year: EN: Entry-level/ Junior, MI: Mid-level/ Intermediate, SE: Senior-level / Expert, EX: Executive-level / Director
# employment_type: Part-time: PT, Full-time: FT, Contract: CT, Freelance: FL
# job_title
# salary
# salary_currency 
# salary_in_usd
# employee_residence: 
# remote_ratio: 0: No remote work (less than 20%), 50: Partially remote/hybrid, 100: Fully remote (more than 80%)
# company_location: 
# company_size: S: less than 50 employees (small), M: 50 to 250 employees (medium), L: more than 250 employees (large)
```

### No.of People by Experience Level by Year

```python
# Create a column chart with the y-axis showing the amount of experience_level
g = sns.catplot(
    data=salaries, kind="count",  # Note kind="count" to show quantity
    x="experience_level", hue="work_year",
    palette="dark", alpha=.6, height=6
)

g.despine(left=True)
g.set_axis_labels("Experience Levels", "No. of People")  # set title for axis
g.legend.set_title("Work Year")  # Set title for legend
g.fig.suptitle('No. of People by Experience Level by Year', y=1.02) # Set title for chart

plt.xticks(rotation=0)  # (Optional) Rotate the labels on the x-axis for easier reading

plt.show()  # Show chart
```

```python
## Senior Level from 2022 to 2023 increased sinificantly 
```

```python
# Creat chart a column chart with the axis y showing the amount of experience_level
g = sns.catplot(
    data=salaries, kind="count",  # Note kind="count" to show quantity
    x="employment_type", hue="work_year",
    palette="dark", alpha=.6, height=6
)

g.despine(left=True)
g.set_axis_labels("Employment Types", "No. of People")  # Set lable for axis
g.legend.set_title("Work Year")  # Set title for legend
g.fig.suptitle('No. of People by Employment Types by Year', y=1.02) # Set title for chart

plt.xticks(rotation=0)  # (Optional) Rotate the labels on the x-axis for easier reading

plt.show()  # Show chart
```

### Top 10 Job Titles 

```python
# Count the Top 10 job titles having the most number people
top_10_job_titles = salaries['job_title'].value_counts().head(10)

# Choose horizontal bar
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_job_titles, y=top_10_job_titles.index, palette='viridis')

# Set title and label
plt.title('Top 10 Job Titles')
plt.xlabel('No. of People')

plt.show()
```

```python
# Top 10 job titles: Data Engineer, Data Scientist, Data Analyst,...
```

### Top 10 Highest Average Salaries by Job Title

```python
# Calculate average salary_in_usd for each of 'job_title' and choose top 10 
top_10_job_titles = salaries.groupby('job_title')['salary_in_usd'].mean().round(1).nlargest(10)

# Set horizontal bar
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_job_titles.values, y=top_10_job_titles.index, palette='viridis')

# Set title and label
plt.title('Top 10 Highest Average Salaries by Job Title')
plt.xlabel('Average Salary (USD)')

plt.show()
```
